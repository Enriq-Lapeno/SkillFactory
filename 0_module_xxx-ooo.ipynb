```python
#функция которая красиво печатает нашу консоль
def prncons(s1,s2,s3):
    print('  0 1 2')
    print('1 '+s1[0]+' '+s1[1]+' '+s1[2])
    print('2 '+s2[0]+' '+s2[1]+' '+s2[2])
    print('3 '+s3[0]+' '+s3[1]+' '+s3[2])

s1='---'
s2='---'
s3='---'

prncons(s1,s2,s3)
```
