Итак, мы имеем:

        Игру «Крестики-нолики».
        Консоль, куда будет выводиться ход игры. Тут делать красиво мы умеем с помощью форматированных строк.
        Неутолимое желание написать что-то реальное своими руками.

Размер поля предполагается равным 3x3.

```python
#функция которая красиво печатает нашу консоль (игровое поле)
# в качестве параметра передаём список list значений 9ти ячеек
def prnconsol(pl):
    print('  0 1 2')
    print('0 '+pl[0]+' '+pl[1]+' '+pl[2])
    print('1 '+pl[3]+' '+pl[4]+' '+pl[5])
    print('2 '+pl[6]+' '+pl[7]+' '+pl[8])

#далее нам нужна функция ТвойХод, которая будет понимать от пользователя куда он ходит
# а так же проверять, не занятя ли уже эта ячейка
# формат ввода XY, где X - номер строки от 0 до 2, и Y - столбца от 0 до 2
# в качестве параметра передаётся значение 'x' или 'o'
def u_turn(x_o):
    celldict = {'00':0, '01':1, '02':2, '10':3, '11':4, '12':5, '20':6, '21':7, '22':8}
    valid = False
    while not valid:
        numcell = input("куда ставим поставим " + x_o+"? ")
        try:
            numcell in {'00', '01', '02', '10', '11', '12', '20', '21', '22'}
        except:
            print ("Вы уверены, что ввели правильно координаты ячейки?")
            continue
        if numcell in {'00', '01', '02', '10', '11', '12', '20', '21', '22'}:
            if xo[celldict[numcell]] == '-':
                xo[celldict[numcell]] = x_o
                valid = True
            else:
                print ("Упс! но это поле уже занято.. Укажите другое")
        else:
            print ("Введите координаты ячейки в формате XY, где X и Y числа от 0 до 2")

#Ещё нам нужна будет функция проверки игрового поля. Назовем эту функцию check_
#в качесткве параметра передаём наш список текущещих значений игрового поля
#но для начала создадим win_ - кортеж (tuple) с выйгрышными значениями 
def check_(pl):
    win_=((0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6))
    for i in win_:
        if pl[i[0]] == pl[i[1]] == pl[i[2]] != '-':
            return pl[i[0]]
    return False
#в случае победы мы видим, что функция возвращает нам значение того кто выйграл 'x' или 'o'

#Осталось создать функцию xo_game, в которой мы соберем вместе все описанные функции.
#Мы ждем когда переменная counter станет больше 4 для того, 
#чтобы избежать заведомо ненужного вызова функции check_ (до пятого хода никто точно не может выиграть). 
#Переменная tmp была создана опять же для того, чтобы лишний раз не вызывать функцию check_, 
#мы просто "запоминаем" ее значение и при необходимости используем.
def xo_game(xo):
    counter = 0
    winer = False
    while not winer:
        prnconsol(xo)
        if counter % 2 == 0:
            u_turn("x")
        else:
            u_turn("o")
        counter += 1
        if counter > 4:
            tmp = check_(xo)
            if tmp:
                print ("Поздравляем! ", tmp, "выиграл!")
                winer = True
                break
        if counter == 9:
            print ("Ничья!")
            break
    prnconsol(xo)
```

```python
xo=['-','-','-','-','-','-','-','-','-']
xo_game(xo)
```
